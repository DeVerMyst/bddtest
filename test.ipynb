```python
from sqlalchemy import create_engine, text
import pandas as pd
import sys
import os 
import datetime # Nécessaire pour l'insertion de la date/heure

# --- Configuration et Connexion ---
DB_FILE = "sakila_master.db"

# Vérification du fichier de base de données
if not os.path.exists(DB_FILE):
    print(f"\nERREUR FATALE : Le fichier de base de données '{DB_FILE}' est introuvable.")
    sys.exit(1)
    
engine = create_engine(f"sqlite:///{DB_FILE}")

# --- Fonction d'Exécution Générique ---
def execute_query(query_title, sql_query, select_verification_query=None):
    """
    Exécute une requête SQL. Gère les SELECT (lecture) et les DML (modification/insertion/suppression).
    """
    print("=" * 70)
    print(f"[{query_title}]")
    print("-" * 70)
    
    try:
        # engine.begin() gère l'ouverture de connexion et le commit/rollback automatique
        with engine.begin() as connection:
            result = connection.execute(text(sql_query))
            
            # Déterminer si c'est une requête de modification de données (DML)
            query_type = sql_query.strip().upper().split()[0]
            
            if query_type in ('INSERT', 'UPDATE', 'DELETE'):
                rows_affected = result.rowcount
                print(f"✅ Opération {query_type} réussie. Lignes affectées : {rows_affected}")

                # Vérification des Changements si une requête SELECT est fournie
                if select_verification_query:
                    print("\n--- Vérification du changement ---")
                    verification_result = connection.execute(text(select_verification_query))
                    column_names = list(verification_result.keys())
                    data = verification_result.fetchall()
                    df = pd.DataFrame(data, columns=column_names)
                    if df.empty and query_type != 'DELETE':
                        print("La requête de vérification n'a retourné aucun résultat (problème de ciblage?).")
                    else:
                        print(df.to_markdown(index=False))
                        
            # Gestion des Requêtes SELECT (Lecture et Agrégation)
            else:
                column_names = list(result.keys())
                data = result.fetchall()
                df = pd.DataFrame(data, columns=column_names)
                if not data:
                    print("Aucun résultat trouvé.")
                    return
                print(df.to_markdown(index=False))
        
    except Exception as e:
        print(f"❌ Erreur d'exécution SQL pour {query_title}: {sql_query[:50]}...")
        print(f"Détails de l'erreur : {e}")
        
    print("=" * 70 + "\n")


# --- Dictionnaire des Requêtes et Exécution ---

if __name__ == "__main__":
    
    # Prépare l'heure actuelle pour l'insertion
    NOW_STR = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    # --- 1. NETTOYAGE PRÉALABLE (ID 201 ciblé) ---
    CLEANUP_BEFORE_SQL = """
    DELETE FROM actor 
    WHERE actor_id = 201;
    """
    execute_query("Nettoyage Préalable : Suppression de l'acteur test (ID 201)", CLEANUP_BEFORE_SQL)

    # -----------------------------------------------------
    # EXERCICE 1 : SELECT, Alias, ORDER BY, LIMIT (Table film)
    # -----------------------------------------------------
    EX1_SQL = """
    SELECT title AS "Titre du Film"
    FROM film
    ORDER BY title ASC
    LIMIT 5;
    """
    execute_query("Ex. 1 : Affichage, Alias, Ordre et Limite (5 films)", EX1_SQL)


    # -----------------------------------------------------
    # EXERCICE 2 : Fonctions d'Agrégation (Table film)
    # -----------------------------------------------------
    EX2_SQL = """
    SELECT
        AVG(length) AS "Durée Moyenne (min)",
        MIN(length) AS "Durée Min (min)",
        MAX(length) AS "Durée Max (min)",
        COUNT(length) AS "Nombre de Films",
        SUM(length) AS "Somme des Durées (min)"
    FROM film;
    """
    execute_query("Ex. 2 : Agrégation (Moyenne, Min, Max, etc.)", EX2_SQL)


    # -----------------------------------------------------
    # EXERCICE 3 : INSERT (Table ACTOR - ID 201 + LAST_UPDATE)
    # -----------------------------------------------------
    # Injection de la variable Python NOW_STR dans la requête F-string
    EX3_INSERT_SQL = f"""
    INSERT INTO actor (actor_id, first_name, last_name, last_update)
    VALUES (201, 'JEAN-CLAUDE', 'VANDAMME', '{NOW_STR}');
    """
    EX3_VERIF_SQL = """
    SELECT actor_id, first_name, last_name, last_update
    FROM actor
    WHERE actor_id = 201;
    """
    execute_query("Ex. 3 : Insertion d'un nouvel acteur (ID 201)", EX3_INSERT_SQL, EX3_VERIF_SQL)


    # -----------------------------------------------------
    # EXERCICE 4 : UPDATE (Table ACTOR - Cible ID 201)
    # -----------------------------------------------------
    EX4_UPDATE_SQL = """
    UPDATE actor
    SET first_name = 'JC'
    WHERE actor_id = 201; 
    """
    EX4_VERIF_SQL = """
    SELECT actor_id, first_name, last_name, last_update
    FROM actor
    WHERE actor_id = 201;
    """
    execute_query("Ex. 4 : Modification du prénom de l'acteur (ID 201)", EX4_UPDATE_SQL, EX4_VERIF_SQL)
    
    print("--- Fin des exercices de SQL Pur ---")
```

```python
from sqlalchemy import create_engine, text
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os 

# --- Configuration et Connexion ---
DB_FILE = "sakila_master.db"

if not os.path.exists(DB_FILE):
    print(f"\nERREUR : Le fichier de base de données '{DB_FILE}' est introuvable.")
    sys.exit(1)
    
engine = create_engine(f"sqlite:///{DB_FILE}")

def run_sql_in_two_steps(sql_query, title):
    """
    Exécute la requête SQL via SQLAlchemy et passe les résultats bruts à Pandas.
    """
    print("=" * 70)
    print(f"[Exercice : {title}]")
    print("-" * 70)
    
    try:
        # ÉTAPE 1: Exécution SQL via SQLAlchemy
        with engine.connect() as connection:
            result = connection.execute(text(sql_query))
            column_names = list(result.keys())
            raw_data = result.fetchall() # Récupération des résultats bruts
        
        print(f"✅ Étape 1 : Requête exécutée par SQLAlchemy. {len(raw_data)} lignes récupérées.")

        # ÉTAPE 2: Création du DataFrame Pandas à partir des données brutes
        df = pd.DataFrame(raw_data, columns=column_names)
        print(f"✅ Étape 2 : DataFrame Pandas créé.")
        return df
        
    except Exception as e:
        print(f"❌ Erreur lors de l'exécution SQL ou du chargement Pandas : {e}")
        return pd.DataFrame()
```

```python
# --------------------------------------------------------------------
# EXERCICE 1 : Distribution des Taux de Location
# --------------------------------------------------------------------
EX1_TITLE = "1. Distribution (Histplot) - SELECT simple"
EX1_SQL = "SELECT rental_rate FROM film;"

df_rental_rate = run_sql_in_two_steps(EX1_SQL, EX1_TITLE)

if not df_rental_rate.empty:
    # ÉTAPE 3: Visualisation (Seaborn Histplot)
    plt.figure(figsize=(8, 5))
    sns.histplot(
        data=df_rental_rate, 
        x='rental_rate', 
        kde=True,
        bins=5,
        color='skyblue'
    ).set(
        title='Distribution des Taux de Location',
        xlabel='Taux de Location ($)',
        ylabel='Nombre de Films'
    )
    plt.show() 
    print("✅ Étape 3 : Diagramme Histplot généré par Seaborn.")
```

```python
# --------------------------------------------------------------------
# EXERCICE 2 : Agrégation et Comparaison (Barplot)
# --------------------------------------------------------------------
EX2_TITLE = "2. Agrégation et Comparaison (Barplot) - GROUP BY"
EX2_SQL = """
SELECT 
    rating, 
    AVG(length) AS avg_length 
FROM 
    film 
GROUP BY 
    rating;
"""

df_duree_par_rating = run_sql_in_two_steps(EX2_SQL, EX2_TITLE)

if not df_duree_par_rating.empty:
    # ÉTAPE 2 (suite): Tri du DataFrame dans Pandas
    df_duree_par_rating = df_duree_par_rating.sort_values(by='avg_length', ascending=False)
    
    # ÉTAPE 3: Visualisation (Seaborn Bar Plot)
    plt.figure(figsize=(8, 5))
    sns.barplot(
        x='rating', 
        y='avg_length', 
        data=df_duree_par_rating,
        palette='viridis'
    ).set(
        title='Durée Moyenne des Films par Classification (Rating)',
        xlabel='Classification (Rating)',
        ylabel='Durée Moyenne (min)'
    )
    plt.show() 
    print("✅ Étape 3 : Diagramme Barplot généré par Seaborn.")
```

```python
from sqlalchemy import create_engine, text
import pandas as pd
import sys

# --- Vérification de la Dépendance Pandas ---
try:
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', 1000)
    # Tente d'utiliser to_markdown() pour vérifier 'tabulate'
    pd.DataFrame({"A": [1]}).to_markdown()
except ImportError:
    print("Erreur : La librairie 'tabulate' est manquante. Installez-la avec : pip install tabulate")
    sys.exit(1)

# --- Configuration et Connexion ---
DB_FILE = "sakila_master.db"
# Utilise le connecteur sqlite3 intégré
engine = create_engine(f"sqlite:///{DB_FILE}")

def execute_query(query_title, sql_query):
    """Exécute une requête SQL et affiche le résultat avec Pandas."""
    print("=" * 60)
    print(f"[{query_title}]")
    print("-" * 60)
    
    try:
        with engine.connect() as connection:
            result = connection.execute(text(sql_query))
            column_names = list(result.keys())
            data = result.fetchall()

        if not data:
            print("Aucun résultat trouvé pour cette requête.")
            return

        # Affichage avec Pandas et le format Markdown
        df = pd.DataFrame(data, columns=column_names)
        print(df.to_markdown(index=False))
        
    except Exception as e:
        print(f"Erreur d'exécution SQL pour {query_title}: {e}")
        
    print("=" * 60 + "\n")


# --- Dictionnaire des Requêtes ---
QUERIES = {
    # EXERCICE 1 : 2 Tables (customer, rental)
    "Ex. 1 : Jointure Simple (2 Tables) - 10 dernières locations": 
    """
    SELECT
        c.first_name AS "Prénom",
        c.last_name AS "Nom",
        r.rental_date AS "Date Location"
    FROM
        customer AS c
    INNER JOIN
        rental AS r ON c.customer_id = r.customer_id
    ORDER BY
        r.rental_date DESC
    LIMIT 10;
    """,

    # EXERCICE 2 : 3 Tables (rental, inventory, film)
    "Ex. 2 : Jointure à 3 Tables - 5 dernières locations PG-13":
    """
    SELECT
        f.title AS "Titre du Film",
        r.rental_date AS "Date Location"
    FROM
        rental AS r
    INNER JOIN
        inventory AS i ON r.inventory_id = i.inventory_id
    INNER JOIN
        film AS f ON i.film_id = f.film_id
    WHERE
        f.rating = 'PG-13'
    ORDER BY
        r.rental_date DESC
    LIMIT 5;
    """,
    
    # EXERCICE 3 : 4 Tables (actor, film_actor, film, language)
    "Ex. 3 : Jointure à 4 Tables - Acteurs dans des films en Anglais (15 premiers)":
    """
    SELECT DISTINCT
        a.first_name AS "Prénom Acteur",
        a.last_name AS "Nom Acteur"
    FROM
        actor AS a
    INNER JOIN
        film_actor AS fa ON a.actor_id = fa.actor_id
    INNER JOIN
        film AS f ON fa.film_id = f.film_id
    INNER JOIN
        language AS l ON f.language_id = l.language_id
    WHERE
        l.name = 'English'
    ORDER BY
        a.last_name
    LIMIT 15;
    """,
    
    # EXERCICE 4 : 5 Tables (rental, inventory, film, store, staff)
    "Ex. 4 : Jointure à 5 Tables - 5 derniers films loués par le magasin n°2":
    """
    SELECT
        f.title AS "Titre du Film",
        r.rental_date AS "Date Location",
        s.first_name AS "Prénom Employé",
        s.last_name AS "Nom Employé"
    FROM
        rental AS r
    INNER JOIN
        inventory AS i ON r.inventory_id = i.inventory_id
    INNER JOIN
        film AS f ON i.film_id = f.film_id
    INNER JOIN
        store AS st ON i.store_id = st.store_id -- Utilisation de i.store_id pour la cohérence
    INNER JOIN
        staff AS s ON r.staff_id = s.staff_id
    WHERE
        st.store_id = 2
    ORDER BY
        r.rental_date DESC
    LIMIT 5;
    """,
    
    # EXERCICE 5 : 6 Tables (customer, address, city, country, payment)
    "Ex. 5 : Jointure à 6 Tables - Clients de Londres avec > 100.00 $ de paiements":
    """
    SELECT
        c.first_name AS "Prénom Client",
        c.last_name AS "Nom Client",
        SUM(p.amount) AS "Total Dépensé"
    FROM
        customer AS c
    INNER JOIN
        address AS a ON c.address_id = a.address_id
    INNER JOIN
        city AS ci ON a.city_id = ci.city_id
    INNER JOIN
        country AS co ON ci.country_id = co.country_id
    INNER JOIN
        payment AS p ON c.customer_id = p.customer_id
    WHERE
        ci.city = 'London'
    GROUP BY
        c.customer_id
    HAVING
        "Total Dépensé" > 100.00
    ORDER BY
        "Total Dépensé" DESC;
    """
}

# --- Exécution principale ---
if __name__ == "__main__":
    print(f"--- Démarrage des 5 exercices de Jointures sur {DB_FILE} ---")
    
    # Vérification que le fichier de base de données existe
    import os
    if not os.path.exists(DB_FILE):
        print(f"\nERREUR FATALE : Le fichier de base de données '{DB_FILE}' est introuvable.")
        print("Assurez-vous qu'il est dans le même dossier que ce script.")
        sys.exit(1)

    for title, query in QUERIES.items():
        execute_query(title, query)
    
    print("--- Fin des exercices ---")
```
